```python
import csv

ubm_labels = []

with open('/sharedfolder/aapb-ubm/UBM_labels/aapb_ubm_labels_4K_corpus.csv') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        ubm_labels.append(row)
```

```python
%%capture

error_rows = []

for row in ubm_labels[1:]:
    try:
        basename, start_time, duration, label, human_labeler = row
        end_time = float(start_time) + float(duration)
        if label in ['Male', 'Female']:
            out_dir = '/sharedfolder/AAPB_UBM_Audio/' + label
            media_path = '/sharedfolder/ubm_clips_final_July_2017/' + basename + '.wav'
            attk.subclip(media_path, float(start_time), end_time, out_dir)
    except:
        error_rows.append(row)
```

```python
%%capture
## Extract features (MFCCs, deltas, and delta-deltas) from Marco Werman & UBM vowel clips, then write features to CSVs

import csv

os.chdir('/sharedfolder/AAPB_UBM_Audio/')
class_dir_pathname = '/sharedfolder/AAPB_UBM_Audio/'


for dir_name in [item for item in os.listdir('./') if os.path.isdir(item)]:
    print("> Starting " + dir_name)
    try:
        os.chdir(os.path.join(class_dir_pathname, dir_name, '_vowel_clips'))
        try: os.mkdir('../_vowel_mfccs_and_deltas_2048/')
        except: pass
        filenames = [item for item in os.listdir('./') if item[-4:].lower()=='.wav']
        for filename in filenames:
            csv_out_path = '../_vowel_mfccs_and_deltas_2048/' + filename[:-4] + '.mfcc.csv'
            if not os.path.isfile(csv_out_path):
                try:
                    mfccs = attk.get_mfccs_and_deltas(filename, freq_min=100, freq_max=16000, hop_length=2048, n_mfcc=13)
                    if len(mfccs) > 0:
                        with open(csv_out_path, 'w') as fo:
                            csv_writer = csv.writer(fo)
                            csv_writer.writerows(mfccs)  
                except Exception as e:
                    print('FILE ERROR: ' + filename)
                    print(e)
    except Exception as e:
        print('SKIPPING DIRECTORY: ' + dir_name)     ## Skipping class directories for which we didn't extract vowels
        print(e)
```
